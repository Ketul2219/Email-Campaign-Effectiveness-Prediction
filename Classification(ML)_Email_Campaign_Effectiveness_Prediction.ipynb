````python
# Write your code to make your dataset analysis ready.


# Data Wrangling

# 1. Handling Missing Values
# Let's check for missing values again
missing_values = df.isnull().sum()
print("Missing Values Count:")
print(missing_values)

# Example: Impute missing values in numerical columns with mean
numerical_columns = df.select_dtypes(include=np.number).columns
for column in numerical_columns:
    df[column].fillna(df[column].mean(), inplace=True)

# Example: Impute missing values in categorical columns with the most frequent category
categorical_columns = df.select_dtypes(exclude=np.number).columns
for column in categorical_columns:
    df[column].fillna(df[column].mode()[0], inplace=True)


# Example: One-hot encoding for categorical columns
df = pd.get_dummies(df, columns=categorical_columns)
````

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

````python

# code for showing bar percentage
def barPerc(df,xVar,ax):
    '''
    barPerc(): Add percentage for hues to bar plots
    args:
        df: pandas dataframe
        xVar: (string) X variable
        ax: Axes object (for Seaborn Countplot)
    '''
    # 1. how many X categories
    ##   check for NaN and remove
    numX=len([x for x in df[xVar].unique() if x==x])

    # 2. The bars are created in hue order, organize them
    bars = ax.patches
    ## 2a. For each X variable
    for ind in range(numX):
        ## 2b. Get every hue bar
        ##     ex. 7 X categories, 3 hues =>
        ##    [0, 8, 16] are hue bars for 1st X category
        hueBars=bars[ind:][::numX]
        ## 2c. Get the total height (for percentages)
        total = sum([x.get_height() for x in hueBars])

        # 3. Print the percentage on the bars
        for bar in hueBars:
            ax.text(bar.get_x() + bar.get_width()/2.,
                    bar.get_height(),
                    f'{bar.get_height()/total:.0%}',
                    ha="center",va="bottom")
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here:-

- Emails with a higher subject hotness score tend to have a higher word count.

- There is a significant difference in the total number of links between emails that were clicked and emails that were not clicked.

- The total number of images in emails varies significantly across different email campaign types.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here:-

**Null Hypothesis (H0)**: There is no significant relationship between the subject hotness score and the word count in emails.

**Alternate Hypothesis (H1)**: There is a significant positive relationship between the subject hotness score and the word count in emails.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load the dataset
df = pd.read_excel("Ketul - ML Classification - Capstone Project - data_email_campaign.xlsx")

# Perform Pearson correlation test
corr_coef, p_value = stats.pearsonr(df['Subject_Hotness_Score'], df['Word_Count'])

print("Pearson's correlation coefficient:", corr_coef)
print("P-value:", p_value)

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a scatter plot with a regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Subject_Hotness_Score', y='Word_Count', data=df, scatter_kws={'s':20}, line_kws={'color':'red'})
plt.title('Scatter Plot of Subject Hotness Score vs Word Count')
plt.xlabel('Subject Hotness Score')
plt.ylabel('Word Count')

# Display the correlation coefficient and p-value on the plot
plt.text(0.05, 0.95, f"Pearson's r: {corr_coef:.2f}\nP-value: {p_value:.2e}",
         transform=plt.gca().transAxes, fontsize=12, verticalalignment='top')

plt.tight_layout()
plt.show()
````

##### Which statistical test have you done to obtain P-Value?

Answer Here:-

To obtain the p-value, I performed comparing the test statistic to a t-distribution with the calculated degrees of freedom.This t-test approach is a standard method for comparing the means of two groups and obtaining a p-value to test the hypothesis regarding the difference in their means.

##### Why did you choose the specific statistical test?

Answer Here:-

I chose the independent samples t-test (also known as the two-sample t-test) for comparing the mean accuracies of two machine learning models because it is specifically designed to determine if there is a statistically significant difference between the means of two independent groups.

The independent samples t-test is a robust, widely-used statistical method for comparing the means of two independent groups, making it an ideal choice for evaluating and comparing the performance of different machine learning models based on their accuracy scores.


### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here:-

**Null Hypothesis (H0)**: There is no significant difference in the average Subject_Hotness_Score between emails that resulted in a successful campaign (Email_Status = 1) and those that did not result in a successful campaign (Email_Status = 0).

**Alternate Hypothesis (H1)**: There is a significant difference in the average Subject_Hotness_Score between emails that resulted in a successful campaign (Email_Status = 1) and those that did not result in a successful campaign (Email_Status = 0).

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

import pandas as pd
from scipy import stats

# Load the dataset
df = pd.read_excel("Ketul - ML Classification - Capstone Project - data_email_campaign.xlsx")

# Extract the relevant columns
subject_hotness_success = df[df['Email_Status'] == 1]['Subject_Hotness_Score']
subject_hotness_failure = df[df['Email_Status'] == 0]['Subject_Hotness_Score']

# Check for normality
w_success, p_success = stats.shapiro(subject_hotness_success)
w_failure, p_failure = stats.shapiro(subject_hotness_failure)

print(f"Success group - Shapiro-Wilk test: W={w_success}, p-value={p_success}")
print(f"Failure group - Shapiro-Wilk test: W={w_failure}, p-value={p_failure}")

# Perform independent t-test
t_stat, p_value = stats.ttest_ind(subject_hotness_success, subject_hotness_failure, equal_var=False)

print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")

# Interpret the p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in the average Subject_Hotness_Score between successful and unsuccessful email campaigns.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the average Subject_Hotness_Score between successful and unsuccessful email campaigns.")
````

### 7. Dimesionality Reduction

##### Do you think that dimensionality reduction is needed? Explain Why?

Answer Here:-

**Curse of Dimensionality**: As the number of features increases, the volume of the feature space grows exponentially, leading to sparsity and increased computational complexity. Dimensionality reduction can help mitigate the curse of dimensionality by reducing the number of features while preserving most of the information.

**Improved Model Performance**: High-dimensional datasets often suffer from overfitting, where the model learns noise in the data rather than the underlying patterns. Dimensionality reduction techniques such as PCA can help remove redundant and irrelevant features, leading to simpler models that generalize better to unseen data.

**Visualization**: Visualizing high-dimensional data is challenging. Dimensionality reduction techniques transform the data into a lower-dimensional space that can be easily visualized in two or three dimensions, enabling better understanding and interpretation of the data's structure.

**Computational Efficiency**: Simplifying the dataset through dimensionality reduction can lead to faster training and inference times for machine learning models. This is particularly important for large-scale datasets where computational resources are limited.

**Collinearity and Redundancy**: High-dimensional datasets often contain features that are highly correlated or redundant. Dimensionality reduction can help identify and remove such features, improving the model's stability and interpretability.

However, dimensionality reduction is not always necessary and may lead to information loss, especially if the reduced features do not capture the essential characteristics of the data. It's essential to carefully evaluate the trade-offs and consider the specific requirements of the problem before deciding whether to perform dimensionality reduction.

````python
# DImensionality Reduction (If needed)


import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Load the dataset
df = pd.read_excel("Ketul - ML Classification - Capstone Project - data_email_campaign.xlsx")

# Separate features and target variable
X = df.drop(columns=['Email_ID', 'Email_Status'])  # Exclude non-numeric columns if necessary
y = df['Email_Status']

# Preprocessing pipeline for numeric features
numeric_features = X.select_dtypes(include=['float64', 'int64']).columns
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# Preprocessing pipeline for categorical features
categorical_features = X.select_dtypes(include=['object']).columns
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(drop='first'))
])

# Combine preprocessing steps for numeric and categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Apply preprocessing and dimensionality reduction using PCA
pca = PCA(n_components=2)
pca_pipeline = Pipeline(steps=[('preprocessor', preprocessor), ('pca', pca)])
X_pca = pca_pipeline.fit_transform(X)

# Create a DataFrame for the PCA-transformed data
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])

# Print the data after dimensionality reduction
print(pca_df.head())


````

````python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier

# Function to evaluate the model
def evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)

    train_precision = precision_score(y_train, y_train_pred, average='macro')
    test_precision = precision_score(y_test, y_test_pred, average='macro')

    train_recall = recall_score(y_train, y_train_pred, average='macro')
    test_recall = recall_score(y_test, y_test_pred, average='macro')

    train_f1 = f1_score(y_train, y_train_pred, average='macro')
    test_f1 = f1_score(y_test, y_test_pred, average='macro')

    result = {
        'Model': model_name,
        'Train Accuracy': train_accuracy,
        'Test Accuracy': test_accuracy,
        'Train Precision': train_precision,
        'Test Precision': test_precision,
        'Train Recall': train_recall,
        'Test Recall': test_recall,
        'Train F1 Score': train_f1,
        'Test F1 Score': test_f1
    }

    return result

# Train the Random Forest model
rf_default_original = RandomForestClassifier(random_state=42)
rf_default_original.fit(X_train, y_train)

# Evaluate the model
result = evaluate_model(rf_default_original, X_train, y_train, X_test, y_test, 'Random_forest_with_Original_data')
final_dict = {}
final_dict['Random_forest_with_Original_data'] = result

# Print the results
for key, value in final_dict['Random_forest_with_Original_data'].items():
    print(f'{key}: {value}')

# Visualize the evaluation metrics
metrics = ['Train Accuracy', 'Test Accuracy', 'Train Precision', 'Test Precision', 'Train Recall', 'Test Recall', 'Train F1 Score', 'Test F1 Score']
scores = [result[metric] for metric in metrics]

plt.figure(figsize=(12, 6))
sns.barplot(x=metrics, y=scores, palette='viridis')
plt.title('Evaluation Metrics for Random Forest Model')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()
````
